# 5 Análisis de los datos

## 5.1 Lanzamiento por año

```python
# Agrupar por año de lanzamiento, contar la cantidad de juegos lanzados y mostrar las plataformas
games_by_year_release = games_df.groupby("year_of_release").agg({
    'name': 'count', # Contar la cantidad de juegos
    'platform' : lambda x: list(x.unique())  # Obtener lista de plataformas únicas
}).reset_index().rename(columns={'name': 'game_count'})
games_by_year_release
```

```python
# Definir rangos de generaciones
generation_ranges = {
    '1st Gen (1972-1983)': (1972, 1983),
    '2nd Gen (1976-1992)': (1976, 1992),
    '3rd Gen (1983-1998)': (1983, 1998),
    '4th Gen (1987-1996)': (1987, 1996),
    '5th Gen (1993-2002)': (1993, 2002),
    '6th Gen (1998-2006)': (1998, 2006),
    '7th Gen (2005-2013)': (2005, 2013),
    '8th Gen (2012-2020)': (2012, 2020),
}

# Asignar generación basada en year_of_release
def assign_generation(years):
    result = []
    for year in years:
        gen_assigned = 'Unknown'
        for gen, (start, end) in generation_ranges.items():
            if start <= year <= end:
                gen_assigned = gen
                break
        result.append(gen_assigned)
    return result

# Asignar generaciones a la columna 'generation'
games_df['generation'] = assign_generation(games_df['year_of_release'])

# Agrupar por generación y contar juegos
games_by_generation = games_df.groupby('generation').agg({
    'name': 'count'
}).reset_index().rename(columns={'name': 'game_count'})

# Imprimir resultados
games_by_generation
```

```python
# Visualizar los datos

# Crear diccionario de traducción para las generaciones
translation_dict = {
    '1st Gen (1972-1983)': '1ra Gen (1972-1983)',
    '2nd Gen (1976-1992)': '2da Gen (1976-1992)',
    '3rd Gen (1983-1998)': '3ra Gen (1983-1998)',
    '4th Gen (1987-1996)': '4ta Gen (1987-1996)',
    '5th Gen (1993-2002)': '5ta Gen (1993-2002)',
    '6th Gen (1998-2006)': '6ta Gen (1998-2006)',
    '7th Gen (2005-2013)': '7ma Gen (2005-2013)',
    '8th Gen (2012-2020)': '8va Gen (2012-2020)',
    '9th Gen (2017-2025)': '9na Gen (2017-2025)',
    'Unknown': 'Desconocida'
}

# Aplicar traducción a las etiquetas de las generaciones para la gráfica
games_by_generation['generation'] = games_by_generation['generation'].map(translation_dict)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.bar(games_by_generation['generation'], games_by_generation['game_count'], color='skyblue', edgecolor='black')
plt.title('Número de Juegos por Generación de Videojuegos', fontsize=14, pad=15)
plt.xlabel('Generación', fontsize=12)
plt.ylabel('Cantidad de Juegos', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

## 5.2 Ventas por plataforma

```python
# Crear tabla dinámica con ventas totales por año y plataforma
games_pivot_table = pd.pivot_table(
    data=games_df,
    values='total_sales',        # Valores a sumar
    index='year_of_release',     # Filas: años
    columns='platform',          # Columnas: plataformas
    aggfunc='sum',              # Sumar las ventas
    fill_value=0                # ara años/plataformas sin ventas
)

games_pivot_table
```

```python
# Calcular ventas totales por plataforma para identificar las mayores
platform_totals = games_pivot_table.sum().sort_values(ascending=False)
platform_totals
```

```python
top_platforms = platform_totals.index[:5]  # Seleccionar las 5 plataformas con mayores ventas
top_platforms
```

```python
# Filtrar la tabla pivot para incluir solo las plataformas principales
pivot_table_top = games_pivot_table[top_platforms]
pivot_table_top = pivot_table_top.query("index >= 2000 and index <= 2016") #Filtrar los años de interés
pivot_table_top
```

```python
# Crear la gráfica del Top 5 de plataformas
plt.figure(figsize=(12, 6))
for platform in top_platforms:
    plt.plot(pivot_table_top.index, pivot_table_top[platform], marker='o', label=platform)

plt.title('Distribución de Ventas Totales de Videojuegos por Año para las Principales Plataformas', fontsize=14, pad=15)
plt.xlabel('Año de Lanzamiento', fontsize=12)
plt.ylabel('Ventas Totales (millones)', fontsize=12)
plt.legend(title='Plataforma')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

### 5.2.1 Ventas por periodo de 6 años.

```python
# Función para obtener una tabala pivote con las plataformas que tuvieron ventas (>0) durante un periodo de 10 años
def years_period(start, end):
    period_pivot_table = games_pivot_table.query("index >= @start and index <= @end") #Filtrar los años de interés
    platform_sales = period_pivot_table.sum()
    period_pivot_table = period_pivot_table.loc[:, platform_sales > 0]
    return period_pivot_table
```

```python
#Función para graficar las ventas de las plataformas
def graph_platform_sales(platforms_sales_period):
    plt.figure(figsize=(12, 6))
    for platform in platforms_sales_period.columns:
        plt.plot(platforms_sales_period.index, platforms_sales_period[platform], marker='o', label=platform)
    
    plt.title(f'Ventas Totales de Videojuegos por Plataforma durante el perido de {platforms_sales_period.index.min()} a {platforms_sales_period.index.max()})', fontsize=14, pad=15)
    plt.xlabel('Año', fontsize=12)
    plt.ylabel('Ventas Totales (millones)', fontsize=12)
    plt.legend(title='Plataforma', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Mostrar la gráfica
    plt.show()
```

# 6. Pruebas de hipótesis

## 6.1 Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

```python
# Imprimir la hiótesis nula:
print("Hipótesis nula (H₀): La calificación promedio de los usuarios para Xbox One es igual a la calificación promedio de los usuarios para PC.")
```

```python
# Imprimir la hipótesis alternativa:
print("Hipótesis alternativa (H₁): Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.")
```

```python
# Filtrar datos para Xbox One y PC
xbox_one_scores = games_df[games_df['platform'] == 'XOne']['user_score']
pc_scores = games_df[games_df['platform'] == 'PC']['user_score']

# Verificar tamaños de las muestras
print(f"Tamaño de la muestra Xbox One: {len(xbox_one_scores)}")
print(f"Tamaño de la muestra PC: {len(pc_scores)}")
```

```python
# verificar los promedios de las muestras
xbox_one_scores_avg = xbox_one_scores.mean()
pc_scores_avg = pc_scores.mean()
print(f"Promedio de puntajes de usuarios en Xbox One: {xbox_one_scores_avg}")
print(f"Promedio de puntajes de usuarios en PC: {pc_scores_avg}")
```

<div class="alert alert-block alert-info">
<b>Sección agregada para verificar las varianzas y hacer la prueba de levene</b> <a class="tocSkip"></a>

```python
# verificar las varianzas de las muestras
xbox_one_scores_var = xbox_one_scores.var()
pc_scores_var = pc_scores.var()
print(f"Varianza de puntajes de usuarios en Xbox One: {xbox_one_scores_var}")
print(f"Varianza de puntajes de usuarios en PC: {pc_scores_var}")
```

```python
# Prueba de Levene para comparar varianzas
results = stats.levene(xbox_one_scores, pc_scores)
print(f"Valor p de Levene: {results.pvalue}")
```

```python
# Interpretación de Levene
# Nivel de significación
alpha = 0.05
if results.pvalue < alpha:
    print("Las varianzas no son iguales (equal_var=False).")
    equal_var = False
else:
    print("Las varianzas son iguales (equal_var=True).")
    equal_var = True
```

<div class="alert alert-block alert-info">
<b>De la prueba de levene se obtuvo que las varianzas son iguales</b> <a class="tocSkip"></a>

```python
# Prueba t para dos muestras independientes (Welch's t-test)
results = stats.ttest_ind(xbox_one_scores, pc_scores, equal_var = equal_var)

# Imprimir el valor p
print('valor p:', results.pvalue)

# Interpretación
if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula: Hay evidencia significativa de que las calificaciones promedio de usuarios difieren entre Xbox One y PC.")
else:
    print("No rechazamos la hipótesis nula: No hay evidencia suficiente para decir que las calificaciones promedio de usuarios difieren entre Xbox One y PC.")
```

<div style="color: blue;">
La hipótesis nula fue RECHAZADA, dado que hay suficiente evidencia para sostener que los promedios de las calificaciones de los usuarios, en XBOX ONE y PC, son diferentes.
</div>
